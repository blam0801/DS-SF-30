> ## Question 4.  Estimate a simple linear regression between `TV` (explanatory variable) and `Sales` (response variable).

````python
model = smf.ols(formula = 'Sales ~ TV', data =df).fit()

# Response variable comes before the Explanatory Variable 
model.summary()
````

> ## Question 5.  Interpret the coefficients.  Are they significant?  What equation did you derive?

Answer: Sales is significant beacuse tvalues are greater than zero, the p value is less than 5% and the confidence intervals don't include 0.

Sales = 7.03 +0.0474(TV)
For the data we have one hand and with evrything else fixed, for every 1000 spent in TV ads, sales increased by 47. 

> ## Question 6.  How are the residual distributed?  Do they follow the linear modeling assumptions?

````python
# TODO
#60% of the redicual have the found. They have been distributed well and do follow the linear modeling assumptions
````

> ## Question 7.  Estimate two new simple linear regression models: (1) `Sales` against `Radio` and (2) `Sales` against `Newspaper`.

````python
model1 = smf.ols(formula = 'Sales ~ Radio', data =df).fit()

model1.summary()
````

````python
model2 = smf.ols(formula = 'Sales ~ Newspaper', data =df).fit()

model2.summary()
````

> ## Question 8.  What can you conclude of these two new models against the first one?

Answer: The R2 values are much lower while the coffeeicients appear to be significant

There is also relative high values of f values but are still lower than the one for TV

> ## Question 9.  Run a multiple linear regression using `TV`, `Radio`, and `Newspaper` as explanatory variables.

````python
smf.ols(formula = 'Sales ~ TV + Radio + Newspaper', data = df).fit().summary()
````
